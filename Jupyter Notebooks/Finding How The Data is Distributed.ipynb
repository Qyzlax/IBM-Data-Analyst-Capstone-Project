### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nSummary Information:")
print(df.info())

print("\nNumerical Summary:")
print(df.describe())

print("\nCategorical Summary:")
print(df.describe(include='object'))
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
missing_values = df.isna().sum()
missing_percent = (missing_values / len(df) * 100).round(2)

missing_summary = (
    pd.DataFrame({
        'Missing Values': missing_values,
        'Missing Percent': missing_percent
    })
    .sort_values('Missing Values', ascending=False)
)

display(missing_summary.head(20))

df_clean = df.copy()

minimum_non_null = int(len(df_clean) * 0.30)
df_clean = df_clean.dropna(axis=1, thresh=minimum_non_null)

text_columns = df_clean.select_dtypes(include=['object', 'string', 'category']).columns
for column in text_columns:
    if df_clean[column].isna().any():
        mode_values = df_clean[column].mode(dropna=True)
        if not mode_values.empty:
            df_clean[column] = df_clean[column].fillna(mode_values.iloc[0])

numeric_columns = df_clean.select_dtypes(include='number').columns
for column in numeric_columns:
    if df_clean[column].isna().any():
        df_clean[column] = df_clean[column].fillna(df_clean[column].median())

print(f"Original shape: {df.shape}")
print(f"Cleaned shape:  {df_clean.shape}")
print(f"Remaining missing values: {df_clean.isna().sum().sum()}")
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
key_columns = ['Employment', 'JobSat', 'YearsCodePro']

for column in key_columns:
    if column in df.columns:
        print(f"\n{'=' * 50}")
        print(f"Value counts: {column}")
        print(df[column].value_counts(dropna=False).head(15))

fig, axes = plt.subplots(1, 3, figsize=(18, 7))

employment_counts = (
    df['Employment']
    .dropna()
    .str.split(';')
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
    .sort_values()
)
employment_counts.plot(kind='barh', ax=axes[0])
axes[0].set_title('Top Employment Types')
axes[0].set_xlabel('Number of Respondents')
axes[0].set_ylabel('')

job_sat_counts = df['JobSat'].dropna().value_counts().sort_index()
job_sat_counts.plot(kind='bar', ax=axes[1])
axes[1].set_title('Job Satisfaction Distribution')
axes[1].set_xlabel('Job Satisfaction Score')
axes[1].set_ylabel('Number of Respondents')
axes[1].tick_params(axis='x', rotation=0)

experience_counts = (
    df['YearsCodePro']
    .dropna()
    .value_counts()
    .head(12)
    .sort_values()
)
experience_counts.plot(kind='barh', ax=axes[2])
axes[2].set_title('Most Common Professional Experience Responses')
axes[2].set_xlabel('Number of Respondents')
axes[2].set_ylabel('')

for axis in axes:
    axis.grid(axis='x', alpha=0.25)

plt.tight_layout()
plt.show()
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
jobsat_counts = df['JobSat'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    jobsat_counts,
    labels=jobsat_counts.index,
    autopct='%1.1f%%'
)
plt.title('Job Satisfaction Distribution')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
from collections import Counter

worked = Counter()
wanted = Counter()

# Languages used
for langs in df['LanguageHaveWorkedWith'].dropna():
    for lang in langs.split(';'):
        worked[lang] += 1

# Languages desired
for langs in df['LanguageWantToWorkWith'].dropna():
    for lang in langs.split(';'):
        wanted[lang] += 1

lang_df = pd.DataFrame({
    'WorkedWith': pd.Series(worked),
    'WantToWorkWith': pd.Series(wanted)
}).fillna(0)

lang_df = lang_df.sort_values(
    'WorkedWith',
    ascending=False
).head(15)

print(lang_df)

lang_df.plot(
    kind='bar',
    figsize=(12,6)
)

plt.title('Programming Languages: Current vs Desired')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
top_countries = df["Country"].value_counts().head(15)

# Shorten long country names
top_countries.index = top_countries.index.str.replace(
    "United Kingdom of Great Britain and Northern Ireland",
    "United Kingdom"
)

plt.figure(figsize=(12, 6))

sns.barplot(
    x=top_countries.index,
    y=top_countries.values,
    hue=top_countries.index,
    palette="viridis",
    legend=False
)

plt.title("Top 15 Countries of Respondents")
plt.xlabel("Country")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

correlation_data = df[['JobSat', 'YearsCodePro']].copy()

correlation_data['YearsCodeProNumeric'] = (
    correlation_data['YearsCodePro']
    .replace({
        'Less than 1 year': 0.5,
        'More than 50 years': 51
    })
    .pipe(pd.to_numeric, errors='coerce')
)

correlation_data['JobSat'] = pd.to_numeric(
    correlation_data['JobSat'],
    errors='coerce'
)

correlation_data = correlation_data.dropna(
    subset=['JobSat', 'YearsCodeProNumeric']
)

pearson_correlation = correlation_data[
    ['JobSat', 'YearsCodeProNumeric']
].corr(method='pearson').iloc[0, 1]

spearman_correlation = correlation_data[
    ['JobSat', 'YearsCodeProNumeric']
].corr(method='spearman').iloc[0, 1]

print(f"Pearson correlation:  {pearson_correlation:.3f}")
print(f"Spearman correlation: {spearman_correlation:.3f}")
print(f"Number of complete responses: {len(correlation_data):,}")

experience_summary = (
    correlation_data
    .groupby('YearsCodeProNumeric', as_index=False)
    .agg(
        MedianJobSat=('JobSat', 'median'),
        Respondents=('JobSat', 'size')
    )
    .query('Respondents >= 20')
)

plt.figure(figsize=(11, 6))
sns.lineplot(
    data=experience_summary,
    x='YearsCodeProNumeric',
    y='MedianJobSat',
    marker='o'
)
plt.title('Median Job Satisfaction by Professional Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction')
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here

employment_education = (
    df[['Employment', 'EdLevel']]
    .dropna()
    .assign(Employment=lambda data: data['Employment'].str.split(';'))
    .explode('Employment')
    .reset_index(drop=True)
)

employment_education['Employment'] = (
    employment_education['Employment'].str.strip()
)

top_employment_types = (
    employment_education['Employment']
    .value_counts()
    .head(8)
    .index
)

employment_education = employment_education[
    employment_education['Employment'].isin(top_employment_types)
].reset_index(drop=True)

employment_education_table = pd.crosstab(
    index=employment_education['EdLevel'].to_numpy(),
    columns=employment_education['Employment'].to_numpy()
)

employment_education_table.index.name = 'Education Level'
employment_education_table.columns.name = 'Employment Type'

display(employment_education_table)

employment_education_percent = (
    employment_education_table
    .div(employment_education_table.sum(axis=1), axis=0)
    .mul(100)
)

employment_education_percent.plot(
    kind='barh',
    stacked=True,
    figsize=(13, 8)
)

plt.title('Employment Composition within Each Education Level')
plt.xlabel('Percentage of Respondents')
plt.ylabel('Education Level')
plt.legend(
    title='Employment Type',
    bbox_to_anchor=(1.02, 1),
    loc='upper left'
)
plt.tight_layout()
plt.show()
```
