The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.info()
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Task 3: Clean heatmap of missing-value percentages by column
missing_pct = (df.isna().mean() * 100).sort_values(ascending=False)
missing_pct = missing_pct[missing_pct > 0]

plt.figure(figsize=(8, max(3, 0.35 * len(missing_pct))))
sns.heatmap(
    missing_pct.to_frame(name='Missing (%)'),
    annot=True,
    fmt='.1f',
    cmap='YlOrRd',
    vmin=0,
    vmax=100,
    linewidths=0.5,
    cbar_kws={'label': 'Missing values (%)'}
)
plt.title('Missing Values by Column', pad=12)
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_employment = df['Employment'].isnull().sum()
print("Number of missing values in Employment:", missing_employment)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
majority_value = df['Employment'].mode()[0]
print("Most frequent value in Employment:", majority_value)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df['Employment'] = df['Employment'].fillna(majority_value)

print("Missing values after imputation:",
      df['Employment'].isnull().sum())
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
employment_counts = df['Employment'].value_counts().head(10).sort_values()

plt.figure(figsize=(9, 5.5))
ax = sns.barplot(
    x=employment_counts.values,
    y=employment_counts.index,
    orient='h'
)

ax.bar_label(ax.containers[0], padding=3, fmt='%d')
ax.set_title('Top Employment Categories After Imputation', pad=12)
ax.set_xlabel('Number of respondents')
ax.set_ylabel('')
ax.grid(axis='x', alpha=0.25)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
