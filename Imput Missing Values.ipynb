### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
duplicate_rows = df[df.duplicated()]
print("Number of duplicate rows:", duplicate_rows.shape[0])
print(duplicate_rows)
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
df = df.drop_duplicates()

print("Shape after removing duplicates:", df.shape)
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()

print("Missing values in each column:")
print(missing_values)
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
missing_remotework = df["RemoteWork"].isnull().sum()

print("Missing values in RemoteWork:", missing_remotework)
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
remotework_counts = df["RemoteWork"].value_counts(dropna=False)

print("Value counts for RemoteWork:")
print(remotework_counts)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
majority_value = df["RemoteWork"].mode()[0]

print("Majority value in RemoteWork:", majority_value)
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
df["RemoteWork"] = df["RemoteWork"].fillna(majority_value)

print("Missing values after imputation:",
      df["RemoteWork"].isnull().sum())
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
## Write your code here
compensation_cols = [
    col for col in df.columns
    if any(keyword in col.lower()
           for keyword in ["comp", "salary", "pay", "income"])
]

print("Compensation-related columns:")
print(compensation_cols)

for col in compensation_cols:
    print(f"Distribution statistics for {col}:")
    print(df[col].describe())

    if df[col].dtype == "object":
        print("Value counts:")
        print(df[col].value_counts().head())
```
