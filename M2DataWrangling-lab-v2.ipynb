#### Step 1: Import the necessary module.


### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.count(),
    'Missing Values': df.isnull().sum()
})

print("Dataset Summary:")
print(summary)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
print("Basic Statistics:")
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
print("Unique Country Values:")
print(df['Country'].unique())

country_clean = df['Country'].astype(str).str.strip()

print("\nPotential Inconsistent Country Entries:")
print(country_clean.value_counts().tail(20))
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
country_mapping = {
    "USA": "United States",
    "U.S.A.": "United States",
    "United States of America": "United States",
    "UK": "United Kingdom",
    "England": "United Kingdom"
}

df['Country'] = df['Country'].replace(country_mapping)

edlevel_mapping = {
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Bachelor",
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Master",
    "Professional degree (JD, MD, etc.)": "Professional Degree",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "Secondary School"
}

df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
employment_encoded = pd.get_dummies(
    df['Employment'],
    prefix='Employment',
    dummy_na=False
)

df = pd.concat([df, employment_encoded], axis=1)

print("One-Hot Encoded Employment:")
print(employment_encoded.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values = df.isnull().sum().sort_values(ascending=False)

print("\nColumns with Highest Missing Values:")
print(missing_values.head(20))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
median_salary = df['ConvertedCompYearly'].median()

df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(
    median_salary
)

print("\nMissing values in ConvertedCompYearly after imputation:")
print(df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
most_frequent_remote = df['RemoteWork'].mode()[0]

df['RemoteWork'] = df['RemoteWork'].fillna(
    most_frequent_remote
)

print("\nMissing values in RemoteWork after imputation:")
print(df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
min_value = df['ConvertedCompYearly'].min()
max_value = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly_Scaled'] = (
    (df['ConvertedCompYearly'] - min_value)
    / (max_value - min_value)
)

print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Scaled']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearly_Log'] = np.log1p(
    df['ConvertedCompYearly']
)

print("\nLog Transformed Salary:")
print(df[['ConvertedCompYearly',
          'ConvertedCompYearly_Log']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
def experience_level(years):
    if pd.isna(years):
        return np.nan

    if years == "Less than 1 year":
        return "Beginner"

    if years == "More than 50 years":
        return "Expert"

    try:
        years = float(years)

        if years < 3:
            return "Beginner"
        elif years < 7:
            return "Intermediate"
        elif years < 15:
            return "Advanced"
        else:
            return "Expert"

    except:
        return np.nan

df['ExperienceLevel'] = df['YearsCodePro'].apply(
    experience_level
)

print("Experience Level Distribution:")
print(df['ExperienceLevel'].value_counts())
```
